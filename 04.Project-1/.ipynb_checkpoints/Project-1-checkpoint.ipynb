### 3.2. Продажи по платформам

Сгруппируем данные по платформам, отбросим все лишние столбцы

```python
groupby_platform_sales = df.groupby(['platform'])['all_sales'].sum().sort_values(ascending = False).reset_index()
groupby_platform_sales.head(3)
```

Для отображения широко диапазона величин(суммарные продажи), когда значения, которые попадают на график отличаются на много порядков или содержат большой разброс данных, применим логарифмическую шкалу. Зададим ***yscale="log"***

```python
ax=groupby_platform_sales.plot(x='platform', y='all_sales', kind='bar', figsize=(12,6),grid=True)
plt.xlabel('Платформы')
plt.ylabel('Суммарные продажи')
plt.title('Суммарные продажи по всем платформам')
plt.legend('')
ax.set(yscale="log")
plt.show()
```

 Выберем платформы с наибольшими суммарными продажами и построим распределение по годам  
 
 Платформы с наибольшими суммарными продажами  поместили в список **spisok_top10**. Сделаем срез по списку

```python
spisok_top10=['PS2','X360','PS3','Wii','DS','PS','GBA','PS4','PC','3DS']
df_spisok_top10=df.query('platform in @spisok_top10')
```

Сделаем сводную таблицу, чтобы год был индексом, столбцы платформами, а значения - продажами.

```python
df_spisok_top10_srez=df_spisok_top10.pivot_table(index='year_of_release', columns='platform', values ='all_sales',aggfunc='sum')
#data_all=top_all.pivot_table(index='year_of_release', columns='platform', values ='all_sales',aggfunc='sum')
df_spisok_top10_srez.head(3)
```

Удалим строку с нулевым годом

```python
df_spisok_top10_srez = df_spisok_top10_srez.drop([0])
```

```python
plt.figure(figsize=(12,8))
ax=sns.lineplot(data=df_spisok_top10_srez,dashes=False)
plt.legend(title='Платформы:')
plt.title('Распределение выпущенных игр с наибольшими суммарными продажами по годам', fontsize=12)
plt.xlabel("Временной промежуток", fontsize=12)
plt.ylabel("Кол-во выпущенных игр", fontsize=12)
plt.show()
```

Разбор графика:
* Есть платформы, которые уже не продаются(пример: GBA,PC2)
* Есть платформы,которые продаются с 1985 года и по сей день(пример:PC,имеет несколько пиков  продаж)
* Представленные основные платформы  были популярны около 10 лет. Но это рассуждение только по тем платформам, которые уже не продаются

Определим средний актуальный период для всех платформ

```python
data_spisok_all=df.pivot_table(index='year_of_release', columns='platform', values ='all_sales',aggfunc='sum')
data_spisok_all = data_spisok_all.drop([0])
```

```python
# функция для определения кол-во годов, когда продавалась платформа
def data_all_mean(df):
    null=[]
    for columns in df:
        a=len(df.loc[df.loc[:,columns] >0]) 
        null.append(a)  
    return null
```

```python
from statistics import median
asre=data_all_mean(data_spisok_all)
print(' В среднем платформа живет', median(asre),'лет' )
```


Из графика распределение выпущенных игр с наибольшими суммарными продажами по годам мы  определили, что актуальный период  каждой платформы составляет около 10 лет, но на анализе всех платформ за весь период численно подтвердили, что в среднем акутальный период состалвяет 8 лет.

### 3.3.  «Ящик с усами» по суммарным продажам игр в разбивке по платформам

Построим график «ящик с усами» по суммарным  продажам игр в разбивке по платформам.  
Для  наглядности построения, используем аргумент **order** в функциях sns.boxplot, чтобы упорядочить "boxes"

```python
df_from2014_order = df_from2014.groupby('platform')['all_sales'].agg('median').sort_values(ascending=False).index
display(df_from2014_order)
```

```python
plt.figure(figsize=(14,8))
ax = sns.boxplot(data=df_from2014, y="all_sales", x='platform',order=df_from2014_order)
ax.set_title('Распределение глобальных продаж по платформам', fontsize=20)
ax.set_xlabel('Игровая платформа', fontsize=15)
ax.set_ylabel('Глобальные продажи', fontsize=15)
ax.set(yscale="log")
plt.show()
```

Boxplot( кроме,PSV,PSP) сильно скошены вправно. Это просто обьяснить - в каждой из этих платформ есть хорошие игры с очень хорошими показателями продаж или игры с очень высокой стоимостью. Как видно из графиков, выбросов (позиции, не входящие в нормальный размах)  достаточное количество для каждого boxplota.     

Среди потенциально прибыльных платформ PS4,WiiU,XOne  медианы очен схожи, что говорит о хорошей конкуренции.



### 3.4 Влияние отзывов на продажи

```python
platform_ps4 = df_from2014.loc[(df_from2014['platform'] == 'PS4')]
sns.set()
ax=sns.pairplot(platform_ps4,x_vars=['critic_score','user_score'], y_vars=['all_sales'],height=5,aspect=1.1,kind="reg")
ax.fig.text(0.3, 1, 'Зависимость продаж PS4  от отызвов с линейной регрессией',fontsize=16)
ax.set(ylabel = 'Суммарные продажи')
plt.show()
```

```python
sns.heatmap(platform_ps4 [['all_sales', 'critic_score', 'user_score']].corr(),annot=True, fmt='.1g')
```

Между суммарными продажами на PS4 и отзывами пользователей корреляции не наблюдается - 0.02

## 5. Проверка гипотез

### 5.1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Примем нулевую гипотезу - средние пользовательские рейтинги платформ Xbox One и PC одинаковые.  
Альтернативная гипотеза - средние пользовательские рейтинги платформ Xbox One и PC различаются.  
Пороговое значение alpha зададим .05  
Используем критерий Стьюдента - st.ttest_ind

```python
XOne = df_from2014.loc[df_from2014.loc[:,'platform'] == 'XOne' ].reset_index(drop=True)
XOne = XOne[pd.notnull(XOne['user_score'])]
XOne = XOne.loc[XOne.loc[:,'user_score'] > 0]
XOne.head(3)
```

```python
PC = df_from2014.loc[df_from2014.loc[:,'platform'] == 'PC' ].reset_index(drop=True)
PC = PC[pd.notnull(PC['user_score'])]
PC = PC.loc[PC.loc[:,'user_score'] > 0]
PC.head(3)
```

```python
print("Дисперсия для XOne", round((XOne['user_score'].var(ddof=1))))
print("Дисперсия для PC", round((PC['user_score'].var(ddof=1))))
```

```python
# фукнция для построения гистограммы и графика плотности распределения оценок пользователей
def plotnost(data):
    sns.distplot(data['user_score'])
    sb.set_style({'axes.grid': True})
    plt.title('Гистограмма и график плотности распределения оценок пользователей', fontsize=9)
    plt.xlabel("Оценки пользователей", fontsize=9)
    plt.ylabel("Плотность", fontsize=9)
    plt.show()
```

```python
plotnost(XOne);plotnost(PC)
```

Распределение близкое к нормальному, дисперсии не равны. Будем использовать **equal_var = True**, для сравнения посмотрим и **equal_var = False**

```python
alpha = 0.05
#1 equal_var = True
results_true_tariff = st.ttest_ind(
    XOne['user_score'],
    PC['user_score'],
    equal_var = True)

print('p-значение: ', results_true_tariff.pvalue)

if (results_true_tariff.pvalue < alpha):
    print("Отвергаем нулевую гипотезу при equal_var = True")
else:
    print("Не получилось отвергнуть нулевую гипотезу при equal_var = True")
    
#2 equal_var = False
results_false_tariff = st.ttest_ind(
   XOne['user_score'],
    PC['user_score'],
    equal_var = False)

print('p-значение: ', results_false_tariff.pvalue)

if (results_false_tariff.pvalue < alpha):
    print("Отвергаем нулевую гипотезу при equal_var = False")
else:
    print("Не получилось отвергнуть нулевую гипотезу при equal_var = False")
```
