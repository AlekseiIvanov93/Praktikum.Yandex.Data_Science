Напишем функции для расчета выручки
>вычтите бесплатный лимит из суммарного количества звонков, сообщений и интернет-трафика; остаток умножьте на значение из тарифного плана; прибавьте абонентскую плату, соответствующую тарифному плану>

Заполним все пропуски нулями

```python
df=df.fillna(0)
```

```python
def money_calls (df):# расчет выручки по звонкам
    tariff = df['tariff']
    if tariff == 'smart':
        # остаток =затраченные минуты- минуты тарифа
        extra_minutes = df['call_minutes'] - tariffs.loc[0,'minutes_included']
        if extra_minutes > 0 :
            call_cost = extra_minutes*tariffs.loc[0,'rub_per_message']
        else :
            call_cost = 0
    else:# аналогично для тарифа ультра
        extra_minutes = df['call_minutes'] - tariffs.loc[1,'minutes_included']
        if extra_minutes > 0 :
            call_cost =  extra_minutes*tariffs.loc[1,'rub_per_message']
        else :
            call_cost = 0
    return call_cost
```

```python
def money_messages (df):# расчет выручки по сообщениям
    tariff = df['tariff']
    if tariff == 'smart':
        # остаток =затраченные сообщ.- сообщ. тарифа
        extra_messages = df['messages_count'] - tariffs.loc[0, 'messages_included']
        if extra_messages > 0 :
            cost_messages =  extra_messages*tariffs.loc[0, 'rub_per_message']
        else :
            cost_messages = 0
    else:# аналогично для тарифа ультра
        extra_messages = df['messages_count'] - tariffs.loc[1, 'messages_included']
        if extra_messages > 0 :
            cost_messages =  extra_messages*tariffs.loc[1, 'rub_per_message']
        else :
            cost_messages = 0
    return cost_messages
```

```python
def money_mb (df):# расчет выручки по интернету
    tariff = df['tariff']
    if tariff == 'smart':
        # остаток =затраченные мб.- мб. тарифа
        extra_mb = df['mb_used'] - tariffs.loc[0, 'mb_per_month_included']

        if extra_mb > 0 :
            cost_mb =  math.ceil(extra_mb/1024)*tariffs.loc[0, 'rub_per_gb']
        else :
            cost_mb = 0
    elif tariff == 'ultra':# аналогично для тарифа ультра
        extra_mb = df['mb_used'] - tariffs.loc[1, 'mb_per_month_included']
       
        if extra_mb > 0 :
            cost_mb =  math.ceil(extra_mb/1024)*tariffs.loc[1, 'rub_per_gb']
        else :
            cost_mb = 0
    return cost_mb
```

```python
def money_tariff (df):# абонентская плата
    tariff = df['tariff']
    
    if tariff == 'smart':
        cost_tariff=550
    else:
         cost_tariff=1950
    return cost_tariff
```

```python
df['cost_calls'] =  df.apply(money_calls, axis = 1)
df['cost_messages'] = df.apply(money_messages, axis = 1)
df['cost_mb'] =  df.apply(money_mb, axis = 1)
df['cost_tariff']=df.apply(money_tariff, axis = 1)
df['all_cost']=df['cost_calls']+df['cost_messages']+df['cost_mb']+df['cost_tariff']
df
```

## 3. Анализ данных

Сгруппируем данные по тарифам, оставим только интересующие столбцы

```python
#группировка по tariff == "smart"
data_smart= df.query('tariff == "smart" ')
data_smart=data_smart.loc[:,['call_minutes','messages_count', 'mb_used', 'all_cost','month']].reset_index()
data_smart['mb_used']=data_smart['mb_used']/1024

#группировка по tariff == "ultra"
data_ultra= df.query('tariff == "ultra"')
data_ultra=data_ultra.loc[:,['call_minutes','messages_count', 'mb_used', 'all_cost','month']].reset_index()
data_ultra['mb_used']=data_ultra['mb_used']/1024

display(data_smart,data_ultra)
```

Создадим пустой df, куда будем заносить информацию по среднеме значению, дисперсии и стандартному отклонению

```python
df_sbor = pd.DataFrame(columns = data_smart.columns)
display(df_sbor)
```

 Расчитаем среднее для параматров  тарифа "Смарт"

```python
smart_mean = []
for column in data_smart:
    mean = data_smart[column].mean()
    smart_mean.append(round(mean,2))
df_sbor.loc[0] = smart_mean   

display(df_sbor.head())
```

Расчитаем дисперсию  для параматров тарифа "Смарт"

```python
smart_variance = []
for column in data_smart:
    variance = np.var(data_smart[column], ddof= 1)
    smart_variance.append(round(variance,2))
df_sbor.loc[1] = smart_variance    

display(df_sbor.head())
```

Расчитаем стандартное отклонение для параматров тарифа "Смарт"

```python
smart_std = []
for column in data_smart:
    standard_deviation = np.std(data_smart[column], ddof= 1)
    smart_std.append(round(standard_deviation,2))
df_sbor.loc[2] = smart_std 

display(df_sbor.head())
```

Расчитаем среднее для параматров тарифа "Ультра"

```python
ultra_mean = []
for column in data_ultra:
    mean = data_ultra[column].mean()
    ultra_mean.append(round(mean,2))
df_sbor.loc[3] = ultra_mean 

display(df_sbor.head())
```

Расчитаем дисперсию для параматров тарифа "Ультра"

```python
ultra_variance = []
for column in data_ultra:
    variance = np.var(data_ultra[column], ddof= 1)
    ultra_variance.append(round(variance,2))
df_sbor.loc[4] = ultra_variance  

display(df_sbor.head())
```

Расчитаем стандартное отклонение для параматров тарифа "Ультра"

```python
ultra_std = []
for column in data_ultra:
    standard_deviation = np.std(data_ultra[column], ddof= 1)
    ultra_std.append(round(standard_deviation,2))
df_sbor.loc[5] = ultra_std 

df_sbor
```

С помощью мультииндекса облагородим таблицу

```python
values = ['mean', 'variance', 'std']
tariffs = ['smart', 'ultra']
ind = pd.MultiIndex.from_product([tariffs, values])

df_information = pd.DataFrame(df_sbor.values, columns = df_sbor.columns, index = ind)
df_information=df_information.loc[:,['call_minutes','messages_count','mb_used','all_cost']]
df_information
```

## 4. Проверка гипотез

**Cредняя выручка пользователей тарифов «Ультра» и «Смарт» различается**

```python
cost_mean_smart =data_smart.groupby('month')['all_cost'].agg(['mean']).round(2)
cost_mean_smart.columns=['all_cost_mean']

cost_mean_ultra =data_ultra.groupby('month')['all_cost'].agg(['mean']).round(2)
cost_mean_ultra.columns=['all_cost_mean']

display(cost_mean_smart.head(),cost_mean_ultra.head())
```

 Примем нулевю гипотезу - средняя выручка пользователей тарифов «Ультра» и «Смарт» не различаются.  

 Альтернативная гипотеза - средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются.

Используем  гипотезу о равенстве средних двух генеральных совокупностей. 
Пороговое значение alpha зададим .05

```python
print("Дисперсия для cost_mean_smart['all_cost_mean']",round((cost_mean_smart['all_cost_mean'].var(ddof=1))))
print("Дисперсия для cost_mean_ultra['all_cost_mean']",round((cost_mean_ultra['all_cost_mean'].var(ddof=1))))
```

Выборки взяты из схожих по параметрам совокупностей, но дисперсии  отличаются на порядок.
Будем использовать **equal_var = True**, для сравнения посмотрим и **equal_var = False**

```python
alpha = 0.05
#1 equal_var = True
results_true_tariff = st.ttest_ind(
    cost_mean_smart['all_cost_mean'],
    cost_mean_ultra['all_cost_mean'],
    equal_var = True)

print('p-значение: ', results_true_tariff.pvalue)

if (results_true_tariff.pvalue < alpha):
    print("Отвергаем нулевую гипотезу при equal_var = True")
else:
    print("Не получилось отвергнуть нулевую гипотезу при equal_var = True")
    
#2 equal_var = False
results_false_tariff = st.ttest_ind(
    cost_mean_smart['all_cost_mean'],
    cost_mean_ultra['all_cost_mean'],
    equal_var = False)

print('p-значение: ', results_false_tariff.pvalue)

if (results_false_tariff.pvalue < alpha):
    print("Отвергаем нулевую гипотезу при equal_var = False")
else:
    print("Не получилось отвергнуть нулевую гипотезу при equal_var = False")
```

**Cредняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов**

```python
moscow = df.groupby(['city', 'month']).mean().round(2)
moscow=moscow.query(' city == "Москва" ')
cost_moscow= moscow.pivot_table(index = 'month' ,  values = 'all_cost' , aggfunc = 'mean')
cost_moscow.columns=['all_cost_mean']
cost_moscow
```

```python
no_moscow = df.groupby(['city', 'month']).mean().round(2)
no_moscow=no_moscow.query(' city != "Москва" ')
cost_no_moscow= no_moscow.pivot_table(index = 'month' ,  values = 'all_cost' , aggfunc = 'mean')
cost_no_moscow.columns=['all_cost_mean']
cost_no_moscow
```

Примем нулевая гипотезу - cредняя выручка пользователей из Москвы не отличается от выручки пользователей из других регионов.  
Альтернативная гипотеза - средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов

Используем  гипотезу о равенстве средних двух генеральных совокупностей.
Пороговое значение alpha зададим .05


```python
print("Дисперсия для cost_moscow['all_cost_mean']",round((cost_moscow['all_cost_mean'].var(ddof=1))))
print("Дисперсия для cost_no_moscow['all_cost_mean']",round((cost_no_moscow['all_cost_mean'].var(ddof=1))))
```

Выборки взяты из схожих по параметрам совокупностей, но дисперсии  отличаются на порядок.
Будем использовать **equal_var = True**, для сравнения посмотрим и **equal_var = False**

```python
alpha = 0.05
#1 qual_var = True
results_true = st.ttest_ind(
    cost_moscow['all_cost_mean'],
    cost_no_moscow['all_cost_mean'],
    equal_var = True)


print('p-значение: ', results_true.pvalue)

if (results_true.pvalue < alpha):
    print("Отвергаем нулевую гипотезу при equal_var = True")
else:
    print("Не получилось отвергнуть нулевую гипотезу при equal_var = True")

#2 equal_var = False
results_false = st.ttest_ind(
    cost_moscow['all_cost_mean'],
    cost_no_moscow['all_cost_mean'],
    equal_var = False)    

print('p-значение: ', results_false.pvalue)

if (results_false.pvalue < alpha):
    print("Отвергаем нулевую гипотезу при equal_var = False")
else:
    print("Не получилось отвергнуть нулевую гипотезу при equal_var = False")
    
```
